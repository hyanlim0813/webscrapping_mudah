```python
import cloudscraper
from bs4 import BeautifulSoup
import pandas as pd
import time

scraper = cloudscraper.create_scraper()  # bypasses some Cloudflare checks
base_url = "https://www.mudah.my/kuala-lumpur/apartment-condominium-for-rent"

all_data = []
max_pages = 5  # adjust how many pages you want to scrape

for page in range(1, max_pages + 1):
    print(f"\nüîÑ Scraping list page {page}")
    url = f"{base_url}?o={page}"
    response = scraper.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    listings = soup.find_all('div', class_='w_100% p_8px_12px_12px d_flex flex-d_column jc_space-between ai_stretch')


    # Collect listing URLs from this page
    listing_urls = []
    for listing in listings:
        link = listing.find('a', href=True)
        relative_url = link['href'] if link else None
        if relative_url:
            listing_urls.append(relative_url)
        print(relative_url)

    # Scrape details from each individual listing
    for relative_url in listing_urls:
        full_url = relative_url
        print(f"üîç Scraping: {full_url}")

        try:
            response = scraper.get(full_url)
            detail_soup = BeautifulSoup(response.content, 'html.parser')

            title = detail_soup.find('span', class_='flex w-full text-xl font-normal mt-[4px] style__StyledSpan-iwjn3z-11 hRihIe')
            title = title.text.strip() if title else "No title"

            price = detail_soup.find('span', class_='text-2xl md:text-[34px] font-bold style__StyledSpan-iwjn3z-11 hRihIe')
            price = price.text.strip() if price else "No price"

            size = detail_soup.find('p', class_='text-base font-open-sans font-bold style__Text-iwjn3z-4 nvSZF')
            size = size.text.strip() if size else "No size"

            all_data.append({
                'url': full_url,
                'title': title,
                'price': price,
                'size': size
            })

            time.sleep(3)  # avoid being blocked

        except Exception as e:
            print(f"‚ùå Failed to scrape {full_url}: {e}")
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
import time

# Set up headless browser options (optional)
chrome_options = Options()
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")
# chrome_options.add_argument("--headless")  # Leave this commented out for now to see the browser

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)

# Define the base URL and the number of pages to scrape
base_url = "https://www.mudah.my/kuala-lumpur/apartment-condominium-for-rent"
max_pages = 5  # Number of pages to scrape
all_data = []  # To store the results

# Loop through pages
for page in range(1, max_pages + 1):
    url = f"{base_url}?o={page}"
    print(f"\nüîÑ Scraping page {page}: {url}")
    driver.get(url)

    try:
        # Wait for listings to load
        WebDriverWait(driver, 15).until(
            EC.presence_of_all_elements_located((By.XPATH, '//div[contains(@class, "flex-d_column") and contains(@class, "d_flex")]'))
        )

        listings = driver.find_elements(By.XPATH, '//div[contains(@class, "flex-d_column") and contains(@class, "d_flex")]')
        print(f"‚úÖ Found {len(listings)} listings on page {page}")

        # Loop through each listing and extract details
        for listing in listings:
            try:
                # Extract the link to the listing
                link = listing.find_element(By.TAG_NAME, "a").get_attribute("href")
                
                # Extract details from the listing page
                listing.get_attribute("href")  # Navigate to listing page if necessary
                driver.get(link)
                
                # Wait for the detail page to load (optional)
                time.sleep(2)
                
                 # Try to extract the details
                title_element = driver.find_elements(By.CSS_SELECTOR, 'span.flex.w-full.text-xl.font-normal')
                title = title_element[0].text.strip() if title_element else "No title"

                price_element = driver.find_elements(By.CSS_SELECTOR, 'span.text-2xl.font-bold')
                price = price_element[0].text.strip() if price_element else "No price"

                size_element = driver.find_elements(By.CSS_SELECTOR, 'p.text-base.font-open-sans.font-bold')
                size = size_element[0].text.strip() if size_element else "No size"

                # Save the data
                all_data.append({
                    'Title': title,
                    'Price': price,
                    'Size': size,
                    'Link': link
                })
                print(f"‚úÖ Extracted: {title} - {price} - {size} - {link}")
            except Exception as e:
                print(f"‚ùå Error extracting details for a listing: {e}")
                continue

    except Exception as e:
        print(f"‚ùå Error on page {page}: {e}")

    # Optional: Add a small delay between pages to avoid being too aggressive
    time.sleep(2)

# Output collected data
print("\n‚úÖ Scraping complete. Extracted data:")
for data in all_data:
    print(data)

driver.quit()
```
